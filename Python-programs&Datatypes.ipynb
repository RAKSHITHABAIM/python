```python
#Armstrong num (number that is equal to the sum of its digits each raised to the power of the number of digits)
#153 → (1³ + 5³ + 3³) = 1 + 125 + 27 = 153

num = int(input("Enter num: "))
temp = num
power = len(str(num))
sum = 0

while temp > 0:
    digit = temp % 10
    sum += digit ** power
    temp = temp // 10

if sum == num:
    print(f"{num} is Armstrong num")
else:
    print(f"{num} is not Armstrong num")

```

```python
import re

def mask_email(email):
    parts = email.split("@")
    if len(parts) != 2:
        return email  # invalid email, return as is
    
    name, domain = parts
    if len(name) <= 2:
        masked_name = name[0] + "*" * (len(name) - 1)
    else:
        masked_name = name[:2] + "*" * (len(name) - 2)
    
    return masked_name + "@" + domain


def mask_phone(phone):
    digits = re.sub(r"\D", "", phone)  # Remove non-digits
    if len(digits) < 4:
        return "*" * len(digits)
    
    return "*" * (len(digits) - 4) + digits[-4:]


# Take input from user
email_input = input("Enter your email: ")
phone_input = input("Enter your phone number: ")

masked_email = mask_email(email_input)
masked_phone = mask_phone(phone_input)

print("\nMasked Email:", masked_email)
print("Masked Phone:", masked_phone)
```

```python
#Merge without using sort()

list_a = [1, 3, 5, 7, 9]
list_b = [2, 4, 6, 8, 10]

i = j = 0
merged = []

while i < len(list_a) and j < len(list_b):
    if list_a[i] < list_b[j]:
        merged.append(list_a[i])
        i += 1
    else:
        merged.append(list_b[j])
        j += 1

merged += list_a[i:]
merged += list_b[j:]

print(merged)
```

```python
#Tic tac toe board

board = [
    ["X", "O", "X"],
    ["O", "X", "O"],
    ["O", "X", "X"]
]


print("Board:")
for row in board:
    print(" | ".join(row))
    print("-" * 9)


x_count = 0
o_count = 0

for row in board:
    x_count += row.count("X")
    o_count += row.count("O")

print("X count:", x_count)
print("O count:", o_count)


winner = None

# Check rows
for row in board:
    if row[0] == row[1] == row[2]:
        winner = row[0]

# Check columns
for col in range(3):
    if board[0][col] == board[1][col] == board[2][col]:
        winner = board[0][col]

# Check diagonals
if board[0][0] == board[1][1] == board[2][2]:
    winner = board[0][0]

if board[0][2] == board[1][1] == board[2][0]:
    winner = board[0][2]

if winner:
    print("Winner:", winner)
else:
    print("No winner")
```

```python
#To Do List

tasks = []
completed = []

while True:
    print("\nTo-Do List Menu")
    print("1. Add task at end")
    print("2. Add task at beginning")
    print("3. Show tasks")
    print("4. Remove task by name")
    print("5. Remove task by position")
    print("6. Mark task as complete")
    print("7. Show completed tasks")
    print("8. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        task = input("Enter task: ")
        tasks.append(task)
        print("Task added.")

    elif choice == "2":
        task = input("Enter task: ")
        tasks.insert(0, task)
        print("Task added at beginning.")

    elif choice == "3":
        if not tasks:
            print("No tasks.")
        else:
            print("Tasks:")
            for i, task in enumerate(tasks, start=1):
                print(f"{i}. {task}")

    elif choice == "4":
        task = input("Enter task name to remove: ")
        if task in tasks:
            tasks.remove(task)
            print("Task removed.")
        else:
            print("Task not found.")

    elif choice == "5":
        pos = int(input("Enter position to remove: ")) - 1
        if 0 <= pos < len(tasks):
            tasks.pop(pos)
            print("Task removed.")
        else:
            print("Invalid position.")

    elif choice == "6":
        task = input("Enter task to mark complete: ")
        if task in tasks:
            tasks.remove(task)
            completed.append(task)
            print("Task marked complete.")
        else:
            print("Task not found.")

    elif choice == "7":
        if not completed:
            print("No completed tasks.")
        else:
            print("Completed tasks:")
            for i, task in enumerate(completed, start=1):
                print(f"{i}. {task}")

    elif choice == "8":
        print("Exiting...")
        break

    else:
        print("Invalid choice. Try again.")
```

```python
# Removing duplicates

my_tuple = (1, 2, 2, 3, 4, 4, 5)
checked = ()

for item in my_tuple:
    if item not in checked:
        checked += (item,)   # must add as tuple

print("Original tuple:", my_tuple)
print("Removed Duplicates:", checked)

#or

print("Using set")

my_tuple = (1, 2, 2, 3, 4, 4, 5)
print(f"Original tuple with duplicates: {my_tuple}")

# Convert to a set to remove duplicates (order is not preserved)
unique_elements_set = set(my_tuple)

# Convert back to a tuple
unique_tuple = tuple(unique_elements_set)
print(f"Tuple with unique elements: {unique_tuple}")
```

```python
#Character Counter

def count_characters(text):
    count={}
    for char in text:
        if char!="":
            count[char]=count.get(char,0)+1     #dictionary.get(key, default_value)
    return count
res=count_characters("hello world")
print(res)

#If key exists → return its value
#If key does NOT exist → return default_value
#If char is already in dictionary → give its current count
#If not → give 0
```
