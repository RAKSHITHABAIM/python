```python
#Here, as the TV and Sales have a higher correlation we will perform the simple linear regression for these variables.
#We first assign the feature variable, `TV`, during this case, to the variable `X` and the response variable, `Sales`, to the variable `y`.
```

```python
X=df[['TV']]
y=df['Sales']
```

```python
# When preparing data for machine learning models, especially with libraries like scikit-learn, the input features (X) are typically expected
# to be in a 2-dimensional format(like a DataFrame with columns representing features and rows representing samples),even if you only have one feature.
# X = advertising['TV']: If you use single square brackets, advertising['TV'] extracts a pandas Series. A Series is a 1-dimensional data structure.
# Scikit-learn models often expect a 2D array for X.
# X = advertising[['TV']]: By using double square brackets, advertising[['TV']] extracts a pandas DataFrame with a single column named 'TV'.
# A DataFrame is a 2-dimensional data structure, which is the format most scikit-learn estimators expect for their input features.
# This ensures that X has the correct shape and is compatible with the LinearRegression model's input requirements.
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=100)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
r_squared = r2_score(y_test, y_pred)
n = len(y_test)  # Number of observations
p = X_test.shape[1]  # Number of predictor variables
adjusted_r_squared = 1 - (1 - r_squared) * (n - 1) / (n - p - 1)

print(f"R-squared: {r_squared}")
print(f"Adjusted R-squared: {adjusted_r_squared}")

plt.figure(figsize=(10, 7))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2) # Diagonal line for perfect prediction
plt.xlabel('Actual sales', fontsize=12)
plt.ylabel('Predicted sales', fontsize=12)
plt.title('Actual vs. Predicted sales', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
