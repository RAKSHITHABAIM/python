```python
# Data Selection & Indexing 
# loc

df = pd.DataFrame({
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Headphones'],
    'Price': [999, 25, 75, 300, 150],
    'Stock': [15, 150, 80, 45, 60],
    'Category': ['Electronics', 'Accessories', 'Accessories', 'Electronics', 'Audio']
}, index=['P001', 'P002', 'P003', 'P004', 'P005'])

# Select the row with index 'P003' 
print("Row with index 'P003'")
print(df.loc['P003'])

# Select rows 'P002' through 'P004' and columns 'Product' and 'Price' only
print("\nRows 'P002' through 'P004' and columns 'Product' and 'Price'")
print(df.loc['P002':'P004',['Product','Price']])

# Select all rows where Category is 'Electronics' using loc with a condition
print("\nRows where Category is 'Electronics'")
print(df.loc[df['Category']=='Electronics'])
```

```python
# iloc
df = pd.DataFrame({
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Headphones'],
    'Price': [999, 25, 75, 300, 150],
    'Stock': [15, 150, 80, 45, 60],
    'Category': ['Electronics', 'Accessories', 'Accessories', 'Electronics', 'Audio']
}, index=['P001', 'P002', 'P003', 'P004', 'P005'])

# Select the first 3 rows using iloc
print("First 3 rows")
print(df.iloc[0:2])

# Select rows at positions 1 and 3 using iloc
print("\nRows at positions 1 and 3")
print(df.iloc[[1,3]])

# Select the last 2 rows and first 2 columns using iloc
print("\nLast 2 rows and first 2 columns")
print(df.iloc[-2:,0:2])

# Explain why df.iloc[1:3] returns rows at positions 1 and 2, not 1, 2, and 3
print("\nIn Pandas, iloc uses Python slicing rules, where the start index is included and the end index is excluded so df.iloc[1:3] returns rows at positions 1 and 2, not 1, 2, and 3")
```

```python
# Filtering with Conditions

df = pd.DataFrame({
    'Name': ['John', 'Sarah', 'Mike', 'Emma', 'Chris', 'Anna'],
    'Department': ['IT', 'HR', 'IT', 'Sales', 'Sales', 'HR'],
    'Salary': [75000, 65000, 80000, 55000, 58000, 62000],
    'Years': [3, 5, 7, 2, 4, 6],
    'Remote': [True, False, True, False, True, False]
})

filtered=df[
    (df['Department']=='IT')&
    (df['Salary']>60000) &
    (df['Remote']==True) &
    (df['Years']>=5)
]
print(filtered)
```

```python
# loc and iloc

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [25, 30, 35, 28, 32],
    'City': ['NYC', 'LA', 'Chicago', 'Houston', 'Phoenix'],
    'Salary': [50000, 60000, 75000, 55000, 62000]
})

# Use loc to select rows where Age > 28, then from those rows use iloc to get the first 2 rows (note: you'll need to chain operations)
res=df.loc[df['Age']>28].iloc[0:2]
print(res)

# Get the salary of the 3rd row using loc and iloc
print("\nSalary of 3rd row using loc")
print(df.loc[2,'Salary'])

print("\nSalary of 3rd row using iloc")
print(df.iloc[2,3])
```

```python
# Multiple Conditions (AND)

df = pd.DataFrame({
    'Name': ['John', 'Sarah', 'Mike', 'Emma', 'Chris', 'Anna'],
    'Department': ['IT', 'HR', 'IT', 'Sales', 'Sales', 'HR'],
    'Salary': [75000, 65000, 80000, 55000, 58000, 62000],
    'Years': [3, 5, 7, 2, 4, 6],
    'Remote': [True, False, True, False, True, False]
})


# In IT department AND working remotely
print("In IT department AND working remotely")
res=df[(df['Department']=='IT')&(df['Remote']==True)]
print(res)

# Salary between 60000 and 75000 AND more than 3 years experience
print("\nSalary between 60000 and 75000 AND more than 3 years experience")
res=df[(df['Salary'].between(60000,75000)) & (df['Years']>3)]
print(res)

# NOT in HR department AND salary above 60000
print("\nNOT in HR department AND salary above 60000")
res=df[(df['Department']!= 'HR') & (df['Salary']>60000)]
print(res)
```

```python
# Multiple Conditions (OR)
# In IT department OR working remotely
print("In IT department OR working remotely")
res=df[(df['Department']=='IT')|(df['Remote']==True)]
print(res)

# Salary less than 60000 OR more than 70000
print("\nSalary less than 60000 OR more than 70000")
res=df[(df['Salary']<60000) | (df['Salary']>70000)]
print(res)

# In Sales department OR has less than 3 years experience
print("\nIn Sales department OR has less than 3 years experience")
res=df[(df['Department']== 'Sales') & (df['Years']<3)]
print(res)
```

```python
# Complex Real-world Filter

df = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=10),
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Product': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B', 'A', 'C'],
    'Quantity': [10, 15, 8, 20, 12, 25, 18, 22, 14, 30],
    'Revenue': [1000, 1500, 800, 2000, 1200, 2500, 1800, 2200, 1400, 3000]
})

# Records from North or South region with revenue > 1200
res=df[(df['Region'].isin(['North','South'])) & (df['Revenue']>1200)]
print(res)

# Product A or B sold in quantities between 10 and 20
print("\nProduct A or B sold in quantities between 10 and 20")
res=df[(df['Product'].isin(['A','B'])) & (df['Quantity'].between(10,20))]
print(res)

# NOT from West region AND revenue is in the top 50% (hint: use quantile)
print("\nNOT from West region AND revenue is in the top 50%")
res=df[(df['Region']!= 'West') & (df['Revenue']>= df['Revenue'].quantile(0.5))]
print(res)
```

```python
#Statistics
# Descriptive Statistics
import pandas as pd
import numpy as np

# Sample data
df = pd.DataFrame({
    'Salary': [50000, 60000, 55000, 80000, 52000, 58000, 120000, 51000],
    'Years_Experience': [1, 3, 2, 5, 2, 3, 10, 1]
})

# 1. Calculate mean, median, std for both columns
stats = df.agg(['mean', 'median', 'std'])
print("Basic statistics:")
print(stats)
print()

# 2. Find correlation between salary and experience
correlation = df['Salary'].corr(df['Years_Experience'])
print(f"Correlation: {correlation:.3f}")
print()

# 3. Identify outliers using IQR method
def find_outliers_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return column[(column < lower_bound) | (column > upper_bound)]

print("Outliers in Salary:")
print(find_outliers_iqr(df['Salary']))
```

```python
#Missing Data
# Handling Missing Values
import pandas as pd
import numpy as np

# Sample data with missing ages
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve', 'Frank'],
    'Age': [15, 16, np.nan, 17, np.nan, 15],
    'Grade_Level': [9, 10, 11, 12, 9, 9]
})

print("Original with missing values:")
print(df)
print(f"\nMissing values: {df['Age'].isnull().sum()}")

# 1. Fill missing ages with median age grouped by Grade_Level
df['Age'] = df.groupby('Grade_Level')['Age'].transform(
    lambda x: x.fillna(x.median())
)

print("\nAfter filling with median by Grade_Level:")
print(df)

# 2. Verify no missing values remain
print(f"\nMissing values after fill: {df['Age'].isnull().sum()}")

# 3. Why median? It's robust to outliers. If one student was 50 years old,
#    the mean would be skewed, but median represents typical student age.
```

```python
#Multiple Aggregations
import pandas as pd

df = pd.DataFrame({
    'Department': ['IT', 'IT', 'HR', 'HR', 'Finance', 'Finance'],
    'Salary': [80000, 120000, 60000, 90000, 70000, 110000],
    'Years_at_Company': [2, 5, 1, 4, 3, 6]
})

# Multiple aggregations with .agg()
result = df.groupby('Department').agg({
    'Salary': ['sum', 'mean', 'count'],
    'Years_at_Company': 'max'
})

# Flatten column names
result.columns = ['Total_Salary', 'Avg_Salary', 'Emp_Count', 'Max_Years']
result = result.reset_index()

print(result)
```

```python
#Merging (SQL-style Joins)
# Create two related DataFrames
products = pd.DataFrame({
    'product_id': [101, 102, 103, 104],
    'product_name': ['Laptop', 'Mouse', 'Keyboard', 'Monitor'],
    'category': ['Electronics', 'Accessories', 'Accessories', 'Electronics']
})

sales_info = pd.DataFrame({
    'product_id': [101, 101, 102, 103, 105],
    'sale_date': ['2024-01-01', '2024-01-02', '2024-01-01', '2024-01-03', '2024-01-04'],
    'quantity': [2, 1, 5, 3, 2],
    'revenue': [2400, 1200, 125, 135, 60]
})

print("Products DataFrame:")
print(products)
print("\nSales DataFrame:")
print(sales_info)
```

```python
#Inner Join
# Inner join - only matching keys
inner_join = pd.merge(products, sales_info, on='product_id', how='inner')
print("\nINNER JOIN (only matching product_ids):")
print(inner_join)
print(f"Rows: {len(inner_join)}")
```

```python
# Left join - all products, even without sales
left_join = pd.merge(products, sales_info, on='product_id', how='left')
print("\nLEFT JOIN (all products):")
print(left_join)
print(f"Rows: {len(left_join)}")
```

```python
# Right join - all sales, even if product missing
right_join = pd.merge(products, sales_info, on='product_id', how='right')
print("\nRIGHT JOIN (all sales):")
print(right_join)
print(f"Rows: {len(right_join)}")
```

```python
# Outer join - everything
outer_join = pd.merge(products, sales_info, on='product_id', how='outer')
print("\nOUTER JOIN (everything):")
print(outer_join)
print(f"Rows: {len(outer_join)}")
```

```python
# Create sample IMDB dataset
movies = pd.DataFrame({
    'movie_id': range(1, 11),
    'title': ['Movie A', 'Movie B', 'Movie C', 'Movie D', 'Movie E', 
              'Movie F', 'Movie G', 'Movie H', 'Movie I', 'Movie J'],
    'year': [2015, 2016, 2015, 2017, 2016, 2015, 2017, 2016, 2018, 2018],
    'genre': ['Action', 'Drama', 'Comedy', 'Action', 'Drama', 
              'Comedy', 'Action', 'Drama', 'Comedy', 'Action'],
    'rating': [7.5, 8.2, 6.9, 7.8, 8.5, 7.2, 8.0, 7.9, 6.8, 7.3],
    'revenue_millions': [50, 30, 20, 80, 45, 25, 90, 35, 15, 70]
})

print("Movies Dataset:")
print(movies)

# Practice 1: Average rating by genre [citation:4]
genre_avg = movies.groupby('genre')['rating'].mean().round(2)
print("\n1. Average rating by genre:")
print(genre_avg)

# Practice 2: Total revenue by year, sorted
revenue_by_year = movies.groupby('year')['revenue_millions'].sum().sort_values(ascending=False)
print("\n2. Total revenue by year (highest first):")
print(revenue_by_year)

# Practice 3: Movies sorted by rating (top first)
top_movies = movies.sort_values('rating', ascending=False)[['title', 'year', 'genre', 'rating']]
print("\n3. Top rated movies:")
print(top_movies)

# Practice 4: Genre with highest average rating [citation:9]
best_genre = movies.groupby('genre')['rating'].mean().idxmax()
best_rating = movies.groupby('genre')['rating'].mean().max()
print(f"\n4. Best genre: {best_genre} ({best_rating:.2f})")

# Practice 5: Year with most movies [citation:4]
year_counts = movies['year'].value_counts()
print(f"\n5. Year with most movies: {year_counts.idxmax()} ({year_counts.max()} movies)")

# Practice 6: Multiple aggregations by genre
genre_stats = movies.groupby('genre').agg(
    avg_rating=('rating', 'mean'),
    total_revenue=('revenue_millions', 'sum'),
    movie_count=('title', 'count'),
    max_rating=('rating', 'max')
).round(2)
print("\n6. Genre statistics:")
print(genre_stats)

# Practice 7: Add rank within year (movies ranked by rating in their year)
movies['rank_in_year'] = movies.groupby('year')['rating'].rank(ascending=False, method='dense')
print("\n7. Rank within year:")
print(movies[['title', 'year', 'rating', 'rank_in_year']].sort_values(['year', 'rank_in_year']))

# Practice 8: Filter groups - genres with avg rating > 7.5
high_rated_genres = genre_stats[genre_stats['avg_rating'] > 7.5]
print("\n8. Genres with average rating > 7.5:")
print(high_rated_genres)
```
